<h2 style="color:#2E86C1;">
📊 SLEEP KPI :
</h2>

```python


sleep_kpi = df.groupby(pd.cut(df['sleep_hours'],
                              bins=[0,5,7,9,12],
                              labels=['Poor Sleep','Normal Sleep','Good Sleep','Oversleep'])
                      )['exam_score'].agg(['mean','count'])

sleep_kpi
```

```python
ax = sleep_kpi['mean'].plot(kind='bar', 
                            color='skyblue', 
                            edgecolor='black', 
                            figsize=(8,5))
plt.title("Exam Score by Sleep Category", fontsize=14)
plt.xlabel("Sleep Category")
plt.ylabel("Average Exam Score")
plt.ylim(0, sleep_kpi['mean'].max() + 10)
plt.xticks(rotation=0)
ax.bar_label(ax.containers[0], labels=[f"count={c}" for c in sleep_kpi['count']], 
             label_type='edge', padding=2)

plt.show()
```

💡 Insight: Academic performance improves with normal and good sleep:

Poor Sleep → Average 12.76

Normal Sleep → Average 16.65

Good Sleep → Average 20.94

Oversleep has the highest average (24.59), but the number of students is very small (218), so its effect may be limited or not representative of all students.

<h2 style="color:#2E86C1;">

```python
social_kpi = df.groupby(
    pd.cut(df['social_media_hours'],
                               bins=[0,2,4,6,10],
                               labels=['Low','Moderate','High','Extreme'])
                       )['productivity_score'].agg(['mean','count'])

social_kpi
```

<h2 style="color:#2E86C1;">
📊 BURNOUT KPI
</h2>

```python
burnout_kpi = df.groupby(
    pd.cut(df['burnout_level'], 
           bins=[0, 30, 50, 70, 100], 
           labels=['Low (0-30)', 'Moderate (30-50)', 'High (50-70)', 'Very High (70-100)'])
)['mental_health_score'].agg(['mean','count'])

burnout_kpi
```

💡 Insight :

Students who use social media infrequently have the highest scores (38.88).

High and moderate use gradually decrease the average scores (37.11 → 35.75).

Extreme use has a slightly higher average than High, but the number of students is very small (95) → possibly a small and unrepresentative sample.

```python
ax = burnout_kpi['mean'].plot(kind='bar', 
                            color='red', 
                            edgecolor='black', 
                            figsize=(8,5))
plt.title("mental_health_score by burnout_level", fontsize=14)
plt.xlabel(" burnout_level")
plt.ylabel("mental_health_score")
plt.ylim(0,  burnout_kpi['mean'].max() + 10)
plt.xticks(rotation=0)
ax.bar_label(ax.containers[0], labels=[f"count={c}" for c in  burnout_kpi['count']], 
             label_type='edge', padding=2)

plt.show()
```

💡 Insight :
    
We would usually expect high levels of fatigue (high to severe) to negatively affect mental health, but in this case:

Severe = 5.81 (above average) but the student count is low

<h2 style="color:#2E86C1;">
📊 PART TIME JOB KPI

💡 Insight :

Part-time work may reduce study time or focus on exams.

<h2 style="color:#2E86C1;">
📊 STUDY KPI
</h2>

```python


study_kpi = df.groupby(pd.cut(df['study_hours'],
                              bins=[0,2,4,6,10],
                              labels=['Low Study','Medium Study','High Study','Intensive'])
                     )['exam_score'].agg(['mean','count'])

study_kpi
```

```python
ax = study_kpi['mean'].plot(kind='bar', color='GREEN', edgecolor='black', figsize=(8,5))

plt.title("Effect of study hours on exam scores", fontsize=14)
plt.xlabel(" study hours")
plt.ylabel("Avg_Exam_Score")
plt.ylim(0, study_kpi['mean'].max() + 5)
plt.xticks(rotation=0)
ax.bar_label(ax.containers[0], labels=[f"count={c}" for c in study_kpi['count']], 
             label_type='edge', padding=2)

plt.show()
```

💡 Insight :

Increased study hours are associated with a significant rise in average grades:

Low → 8.85

Average → 13.98

High → 19.97

Intensive → 27.61

The relationship is clear: the more study hours, the higher the academic performance.
